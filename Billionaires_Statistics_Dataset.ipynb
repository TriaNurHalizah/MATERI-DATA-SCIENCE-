```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df = pd.read_csv("Billionaires_Statistics_Dataset (2).csv")
df.head()
```

```
df.describe(include='all')
df.info()
df['country'].value_counts().head(10)
df['category'].value_counts().head(10)
```

```
# Top 10 negara
df['country'].value_counts().head(10).plot(kind='bar', title='Top 10 Countries with Most Billionaires')
plt.show()

# Gender distribution
sns.countplot(data=df, x='gender')
plt.title('Gender Distribution')
plt.show()

# Scatterplot kekayaan vs usia
sns.scatterplot(data=df, x='age', y='finalWorth', hue='gender')
plt.title('Wealth vs Age by Gender')
plt.show()
```

```
import seaborn as sns
import matplotlib.pyplot as plt
```

```
sns.violinplot(data=df, x="gender", y="finalWorth", palette="pastel")
plt.title("Distribusi Kekayaan: Laki-laki vs Perempuan")
plt.xlabel("Gender")
plt.ylabel("Kekayaan (dalam milyar USD)")
plt.show()
```
